# Exploratory Data Analysis for all Data Roles

## Job Roles to Explore

```python
## Importing Libraries
import ast
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## Loading Data
df=pd.read_csv("C:/Users/sujit/Downloads/data_jobs.csv")
```

```python
# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])  ##converting into datetime type
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
## This line of code converts string representations of Python lists (or other literals) 
# into actual Python objects for a Pandas DataFrame column.
```

```python
df
```

```python
df_plot = df['job_title_short'].value_counts().to_frame()
df_plot
```

```python
sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='job_title_short', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Number of Jobs per Job Title')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

## Countries to Explore

```python
df_plot = df['job_country'].value_counts().to_frame().head(20) ## considering top 20 countries
df_plot
```

```python
sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='job_country', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Number of Jobs per Country')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

## Companies to Explore

```python
df_plot = df['company_name'].value_counts().to_frame()[1:].head(20) 
df_plot
```

```python
sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='company_name', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Number of Jobs per Company')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

## Job Opportunities

```python
dict_column = {
    'job_work_from_home': 'Work from Home Offered',
    'job_no_degree_mention': 'Degree Requirement',
    'job_health_insurance': 'Health Insurance Offered'
}

fig, ax = plt.subplots(1, 3, figsize=(11, 3.5))

for i, (column, title) in enumerate(dict_column.items()):
    ax[i].pie(df[column].value_counts(), labels=['False', 'True'], autopct='%1.1f%%', startangle=90)
    ax[i].set_title(title)

plt.show()
```

## Exploratory Data Analysis for Data Analysts in the India

### Filter for India Data Analyst roles

```python
df_DA_IND = df[(df['job_country'] == 'India') & (df['job_title_short'] == 'Data Analyst')]
```

```python
df_DA_IND
```

### Locations to Explore:

```python
df_plot = df_DA_IND['job_location'].value_counts().head(10).to_frame()

sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='job_location', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Counts of Job Locations for Data Analyst in the India')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

### Job Opportunities

```python
fig, ax = plt.subplots(1, 3, squeeze=False)
fig.set_size_inches((12, 5))

for i, (column, title) in enumerate(dict_column.items()):
    counts = df_DA_IND[column].value_counts()
    labels = counts.index.astype(str).tolist()
    ax[0, i].pie(counts, labels=labels, autopct='%1.1f%%', startangle=90)
    ax[0, i].set_title(title)

plt.show()
```

### Companies to Explore:

```python
df_plot = df_DA_IND['company_name'].value_counts().head(10).to_frame()
df_plot
```

```python
sns.set_theme(style='ticks')
sns.barplot(data=df_plot, x='count', y='company_name', hue='count', palette='dark:b_r', legend=False)
sns.despine()
plt.title('Counts of Companies for Data Analyst in the US')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```
